```python
#Forecast the closing prices for 10 days for stocks on the basis of the historical data. Before forecasting we should test if the historical data has a trend first (have test if historical trend exist first yet).
import yfinance as yf
import numpy as np
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt 
```

```python
#downloading stock information from Yahoo Finance
gold_df = yf.download("GC=F", start='2019-05-31', end='2021-05-31')
btc_df = yf.download("BTC-USD", start='2019-05-31', end='2021-05-31')
```

```python
#Dropping extra columns and renaming columns
gold_df = gold_df.drop(columns = ['Open','High','Low','Adj Close','Volume']).rename(columns = {'Close':'GOLD'})
btc_df = btc_df.drop(columns = ['Open','High','Low','Adj Close','Volume']).rename(columns = {'Close':'BITCOIN'})
```

```python
# Combine all stocks in a single Dataframe
gold_vs_btc_df = pd.concat([gold_df,btc_df], axis="columns", join="inner")
gold_vs_btc_df.sort_index(inplace=True)
gold_vs_btc_df
```

```python
#Reshaping both dataframes
gold_vs_btc_df.shape
```

```python
gold_vs_btc_df["GOLD"].value_counts()
```

```python
#Resample daily stock closing prices by month using ‘end’ convention. Values are assigned to the last month of the period.
y = gold_df.resample("M").mean()
z = btc_df.resample("M").mean()
```

```python
train_gold=y.loc[:'2021-01-31']
print(train_gold.shape)
train_gold.tail()
```

```python
train_btc=z.loc[:'2021-01-31']
print(train_btc.shape)
train_btc.tail()
```

```python
test_gold=y.loc['2021-02-28':]
print(test_gold.shape)
test_gold.tail()
```

```python
test_btc=z.loc['2021-02-28':]
print(test_btc.shape)
test_btc.tail()
```

```python
#Plotting Gold Data 
train_gold.plot(figsize=(10,6), title= 'Average Closing Price', fontsize=14)
test_gold.plot(figsize=(10,6), title= 'Average Closing Price', fontsize=14)
plt.show()
#Need to figure out how to combine gold graph together
```

```python
#Plotting Bitcoin Data 
train_btc.plot(figsize=(10,6), title= 'Average Closing Price', fontsize=14)
test_btc.plot(figsize=(10,6), title= 'Average Closing Price', fontsize=14)
plt.show()
#Need to figure out how to combine gold graph together
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
Exp_Smooth = test_gold.copy()
#smoothing level=alpha
#alpha value should be in the range of 0 to 1.
#values close to 0 indicate that older values are very less imp compared to the recent values
#values close to 1 indicate that older values are equally imp
fit1 = SimpleExpSmoothing(train_gold).fit(smoothing_level=0.01)
Exp_Smooth['SES'] = fit1.forecast(steps=len(test_gold))

train_gold.plot(figsize=(10,6), title= 'Closing Gold Prices', fontsize=14)
test_gold.plot(figsize=(10,6), title= 'Closing Gold Prices', fontsize=14)
Exp_Smooth['SES'].plot(figsize=(10,6), title= 'Closing Gold Prices', fontsize=14)
plt.show()
```

```python
Exp_Smooth = test_btc.copy()
#smoothing level=alpha
#alpha value should be in the range of 0 to 1.
#values close to 0 indicate that older values are very less imp compared to the recent values
#values close to 1 indicate that older values are equally imp
fit2 = SimpleExpSmoothing(train_btc).fit(smoothing_level=0.01)
Exp_Smooth['SES'] = fit2.forecast(steps=len(test_btc))

train_btc.plot(figsize=(10,6), title= 'Closing Bitcoin Prices', fontsize=14)
test_btc.plot(figsize=(10,6), title= 'Closing Bitcoin Prices', fontsize=14)
Exp_Smooth['SES'].plot(figsize=(10,6), title= 'Closing Bitcoin Prices', fontsize=14)
plt.show()
```

```python
#AIC is used to compare different models with same Algo but different data/parameters
#choose the model with lowest AIC value
fit1.aic
```

```python
fit2.aic
```

```python
#RMSE is used to compare different models with same data but different Algos
#choose the model with lowest RMSE value
from sklearn.metrics import mean_squared_error
from math import sqrt
rms_gold = sqrt(mean_squared_error(test_gold, Exp_Smooth.SES))
print(rms_gold)
```

```python
rms_btc = sqrt(mean_squared_error(test_btc, Exp_Smooth.SES))
print(rms_btc)
```

```python
#Gold Holt
from statsmodels.tsa.api import Holt
Holt_gold_df = test_gold.copy()
#soothing slope=beta
fit1 = Holt(train_gold).fit(smoothing_level = 0.01,smoothing_slope = 0.25)
Holt_gold_df['Holt_linear'] = fit1.forecast(steps=len(test_gold))

train_gold.plot(figsize=(10,6), title= 'Closing Gold Prices', fontsize=14)
test_gold.plot(figsize=(10,6), title= 'Closing Gold Prices', fontsize=14)
Holt_gold_df['Holt_linear'].plot(figsize=(10,6), title= 'Closing Gold Prices', fontsize=14)
plt.show()
```

```python
#Bitcoin Holt
Holt_btc_df = test_btc.copy()
#soothing slope=beta
fit2 = Holt(train_btc).fit(smoothing_level = 0.01,smoothing_slope = 0.25)
Holt_btc_df['Holt_linear'] = fit2.forecast(steps=len(test_btc))

train_btc.plot(figsize=(10,6), title= 'Closing Bitcoin Prices', fontsize=14)
test_btc.plot(figsize=(10,6), title= 'Closing Bitcoin Prices', fontsize=14)
Holt_btc_df['Holt_linear'].plot(figsize=(10,6), title= 'Closing Bitcoin Prices', fontsize=14)
plt.show()
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
Holt_Winter_gold_df = test_gold.copy()
#seasonalperiods=gamma
fit1 = ExponentialSmoothing(train_gold ,seasonal_periods=20,
                            trend='add',
                            seasonal='add').fit()
Holt_Winter_gold_df['Holt_Winter'] = fit1.forecast(steps=len(test_gold))

train_gold.plot(figsize=(10,6), title= 'Closing Gold Price', fontsize=14)
test_gold.plot(figsize=(10,6), title= 'Closing Gold Price', fontsize=14)
Holt_Winter_gold_df['Holt_Winter'].plot(figsize=(10,6), title= 'Closing Gold Price', fontsize=14)
plt.show()
```

```python
Holt_Winter_btc_df = test_btc.copy()
#seasonalperiods=gamma
fit2 = ExponentialSmoothing(train_btc ,seasonal_periods=20,
                            trend='add',
                            seasonal='add').fit()
Holt_Winter_btc_df['Holt_Winter'] = fit2.forecast(steps=len(test_btc))

train_btc.plot(figsize=(10,6), title= 'Closing Bitcoin Price', fontsize=14)
test_btc.plot(figsize=(10,6), title= 'Closing Bitcoin Price', fontsize=14)
Holt_Winter_btc_df['Holt_Winter'].plot(figsize=(10,6), title= 'Closing Bitcoin Price', fontsize=14)
plt.show()
```
